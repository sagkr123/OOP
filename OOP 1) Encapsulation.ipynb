```python
# OOP used for standardization to our code
# Solve complex problem's in a simple way
# 4 pillars solve problem
# 1.Encapsulation
# 2.Inheritance
# 3.Abstraction
# 4.Polymorphisom 

# Encapsulation-->First level binding code in single
# Encapsulation is the practice of keeping an object's data hidden and only allowing acces through specific methods
# This helps protect the data and keeps the code organized

class Account:
    bankname = "ICIC"
    branch = "mysore"

    def __init__(self,acc_no,acc_name,balance):
        self.acc_no = acc_no
        self.acc_name = acc_name
        self.balance = balance

    def acc_info(self):
        print(f"Bank Name: {self.bankname}\nBranch: {self.branch}")
        print(f"Account Number:{self.acc_no}\nAccount Name :{self.acc_name}\nBalance:{self.balance}")
a1 = Account(64293421,'Sagar',31002)
a1.acc_info()
```
