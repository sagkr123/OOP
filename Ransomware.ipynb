```python
import numpy as np
from sklearn.ensemble import IsolationForest

normal_traffic = np.array([[10], [12], [9], [11], [10], [13]])

model = IsolationForest(contamination=0.25)
model.fit(normal_traffic)

test_traffic = np.array([[11], [10], [80], [95], [12]])

result = model.predict(test_traffic)

for i, value in enumerate(test_traffic):
    if result[i] == -1:
        print(f"⚠️ Attack Detected: {value[0]} requests/sec")
    else:
        print(f"Normal Traffic: {value[0]} requests/sec")
```
