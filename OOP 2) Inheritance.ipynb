```python
# Inheritance
# Inheritance is an object-oriented programming concept that allows a new class (child class)
# -> to inherit properties and behaviors (methods) from an existing class (parent class). 
# Resusability code

# 1.Single Inheritance
# Single inheritance occurs when a child class inherits from just one parent class, 
# ->allowing it to use that parent's properties and methods.

class A:
    def __init__(self):
        self.a = 100
        self.b = 200

    def m1(self):
        print("I am m1 of class a")

class B(A):
    def __init__(self):
        super().__init__()
        # Using super() allows a child class to access methods and properties from its parent class.
        self.x = 101
        self.y = 201

    def m2(self):
        print("I am m2 of class b")

b_instance = B()
print(b_instance.a)
print(b_instance.b)
print(b_instance.x)
print(b_instance.y)

b_instance.m1()
b_instance.m2()
```

```python
# 2.Multilevel
# Multilevel inheritance is a type of inheritance where a class inherits from another child
# ->class, forming a chain of classes. This allows the child class to access properties and
# methods from both its direct parent and the grandparent class.

class One:
    def __init__(self):
        self.a = 10
        self.b = 20

class Two(One):
    def __init__(self):
        super().__init__()
        self.x= 30
        self.y= 40

class Three(Two):
    def __init__(self):
        super().__init__()
        self.p = 1000
        self.q = 2000

o1 = One()
print(o1.__dict__)
o2 = Two()
print(o2.__dict__)
o3 = Three()
print(o3.__dict__)
```

```python
# 3.Multiple (One class depends on two class)
# Multiple inheritance is a technique in object-oriented programming where a single class derives from two or more base classes, 
# enabling it to inherit and utilize their properties and methods.
class One:
    def __init__(self):
        self.a=10
class Two:
    def __init__(self):
        self.b=20

class Three(One, Two):
    def __init__(self):
        One.__init__(self)
        Two. __init__(self)
        self.p=100
        self.q=200

o3 = Three()
print(o3.__dict__)
```
